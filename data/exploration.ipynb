```python
import json

# Point to your newly created file
file_path = r"C:\Users\helpf\Desktop\DD2477 Project\Github Repo\PodSeek\cleaned_output\elastic_ready.jsonl"

print("🔍 Inspecting the first 2 chunks...\n")

with open(file_path, 'r', encoding='utf-8') as f:
    for i in range(2):
        line = f.readline()
        if not line:
            break
        # Parse the JSON and print it beautifully
        chunk_data = json.loads(line)
        print(json.dumps(chunk_data, indent=4))
        print("-" * 40)
```
